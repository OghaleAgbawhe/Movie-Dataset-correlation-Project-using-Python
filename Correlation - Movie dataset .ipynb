```python
#Import libraries 

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots to be created 

pd.options.mode.chained_assignment = None

#Read in the data
df = pd.read_csv(r'C:\Users\oghal\OneDrive\Desktop\Data analytics portfolio\Python correlation project\movies.csv')
```

```python
#To get a glimpse of the data - top 5 rows 

df.head()
```

```python
#The entire data 

df
```

```python
#Checking for missing data 

#Looping through each column to check

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing,3)))
```

```python
# To eliminate any missing data

df = df.dropna()

#Checking for confirmation 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing,3)))
```

```python
#Data cleaning 

#Data types for all columns 

df.dtypes
```

```python
#Modifying the data type of certain columns - (Budget and Gross columns specifically)

df['budget'] = df['budget'].astype('Int64')

df['gross'] = df['gross'].astype('Int64')


df
```

```python
#Ordering the data by the gross revenue - Highest grossing movies 

df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
df
```

```python
#To enable scrolling through the data 

pd.set_option('display.max_rows',20)
```

```python
#Checking for duplicates and dropping them leaving only unique values 

df['company'].drop_duplicates().sort_values(ascending= False)

#Can also use df.drop_duplicates()
```

```python
'''
Data is now clean. Next is the Correlation analysis
Obtaining the variables mostly correlated to the 'gross' variable 

Hypothesis: 
Budget might highly correlate with gross revenue 
Company might highly correlate with gross revenue 
Votes might highly correlate with gross revenue

'''
```

```python
#1. Budget and gross revenue
#scatterplot of budget vs. gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings (hundreds of millions)')

plt.ylabel('Budget for film (hundreds of millions)')

plt.show()
```

```python
#Plot the regression of Budget vs Gross using seaborn 

sns.regplot(x= 'budget', y= 'gross', data= df, scatter_kws = {"color":"red"}, line_kws={"color":"blue"})

#The line shows positive correlation 
```

```python
#Determining the amount of correlation

#Pearson, Spearman and Kendall correlation types
#df.corr() - gives the default (Pearson)
```

```python
#Pearson correlation 

df.corr(method= 'pearson', numeric_only = True )

#High corerlation exists between Budget and gross 
```

```python
#Spearman correlation

df.corr(method= 'spearman', numeric_only = True)
```

```python
#Kendall correlation 

df.corr(method= 'kendall', numeric_only = True)
```

```python
#Visualizing the correlation matrix 

correlation_matrix = df.corr(method= 'pearson', numeric_only = True)

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#Investigating the company 

df.head()
```

```python
#representing values of the 'company' variable with numerics for further analysis 
#Generally, this step converts non-numeric columns into random numbers for further analysis as shown below 

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
#Organizing the above to easily see the variables with the highest correlation using unstacking - understanding correlation pairs 

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
#Sorted pairs alongside their correlation values 

sorted_pairs = corr_pairs.sort_values()

sorted_pairs

#The below output is still in the correlation type
```

```python
#Indicating the values with high correlation (sorted pairs with correlation values > 0.5)

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

```python
#The hypothesis of the company being significant or influencing a movie's gross earnings is disproven
#Company has low correlation

#Votes and budget have the highest correlation to a movie's gross earnings. 
#Results for the Votes and Budget variables support the hypothesis.

```
